```python
def extend_df(df):
    hold = []
    for row in df.itertuples():
        for year in range(int(row.FROM_YEAR), int(row.TO_YEAR)+1):
            hold.append([row.PERSON_ID, row.DISPLAY_FIRST_LAST, '{}-{}'.format(str(year), str(year + 1)[2:])])
    return pd.DataFrame(hold, columns = ['PERSON_ID', 'PLAYER_NAME', 'SEASON'])
```

```python
%%time
import async_timeout
from itertools import cycle
from time import sleep
import requests
def get_proxies():
    url = 'https://free-proxy-list.net/'
    response = requests.get(url)
    soup2 = BeautifulSoup(response.text, 'html.parser')
    ips = []
    for link in soup2.find_all('tr'):
        try:
            ips.append(':'.join(list(map(lambda x: x.text, link.find_all('td')[:2]))))
        except:
            continue
    return list(filter(lambda x: re.search('^[\d\.\:]+$', x), ips))
    
proxies = get_proxies()
np.random.shuffle(proxies)
proxy_pool = cycle(proxies)


async def fetch(session, player_id, season):
    params = {
        'DateFrom':'',
        'DateTo':'', 
        'GameSegment': '',
        'LastNGames': 0,
        'LeagueID': '00',
        'Location': '',
        'MeasureType': 'Base',
        'Month': 0,
        'OpponentTeamID': 0,
        'Outcome':'',
        'PORound': 0,
        'PaceAdjust': 'N',
        'PerMode': 'Totals',
        'Period': 0,
        'PlayerID': player_id,
        'Season': season,
        'PlusMinus': 'N',
        'Rank': 'N',
        'SeasonSegment': '',
        'SeasonType': 'Regular Season',
        'ShotClockRange': '',
        'VsConference':'', 
        'VsDivision':''
    }
    
    header = {
        'Accept': 'application/json, text/plain, */*',
        'Accept-Encoding': 'gzip, deflate, br',
        'Accept-Language': 'en-US,en;q=0.9',
        'Connection': 'keep-alive',
        'Host': 'stats.nba.com',
        #'Referer': 'https://stats.nba.com/player/202689/',
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/69.0.3497.100 Safari/537.36',
        'x-nba-stats-origin': 'stats',
        'x-nba-stats-token': 'true',
        'X-NewRelic-ID': 'VQECWF5UChAHUlNTBwgBVw=='
    }
    #header['Referer'] = 'https://stats.nba.com/player/{}/boxscores-traditional/'.format(player_id)
    proxy = next(proxy_pool)
    proxy_test = True
    
    test_count = 0
    while(proxy_test):
        try:
            async with session.get('https://stats.nba.com/stats/playergamelogs/', 
                                   proxy='http://' + proxy, 
                                   headers=header, 
                                   params=params,
                                   timeout=3) as response:
                response_content = await response.json()
                proxy_test=False

        except:
            proxy = next(proxy_pool)
            test_count+=1
            await asyncio.sleep(np.random.uniform(0,.5))
        if test_count > 100:
            print(player_id + ' ' + season)
            response_content ={'resultSets':[{'rowSet':['Missed {} {}'.format(player_id, season)]}]}
            proxy_test = False
    return response_content['resultSets'][0]['rowSet']



async def bound_fetch(sem, session, player_id, season):
    # Getter function with semaphore.
    async with sem:
        return await fetch(session, player_id, season)
    
async def run(df_ext):
    tasks = []
    sem = asyncio.Semaphore(100)
    async with aiohttp.ClientSession() as session:
        for row in df_ext.itertuples():
            task = asyncio.ensure_future(bound_fetch(sem, session, row.PERSON_ID, row.SEASON))
            tasks.append(task)

        responses = await asyncio.gather(*tasks)
    return responses
        
# async def main(loop):
#     urls = df.pid.unique()[:100]
 
#     async with aiohttp.ClientSession(loop=loop) as session:
#         tasks = [download_coroutine(session, url) for url in urls]
#         await asyncio.gather(*tasks)

loop = asyncio.get_event_loop()
future = asyncio.ensure_future(run(df_ext.iloc[:300]))
loop.run_until_complete(future)
```
